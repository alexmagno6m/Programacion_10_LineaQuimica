# Actividades interactivas

## Sopa de letras

Encuentra las 20 palabras relacionadas con programacion, la nota sera de acuerdo al tiempo que te gastes.

[Enlace sopa de letras](https://www.educaplay.com/learning-resources/28407555-sopa_de_letras_conceptos_de_programacion.html)

### Obtencion del sello

Escribe en tu cuaderno cuando termines la sopa de letras:

> Competencia: 1
>
> Ambito conceptual: Los algoritmos en la vida común y en la programación.
>
> Objetivo: Reconocer conceptos relacionados a la programacion y seleccionarlos para hallar la solución a la sopa de letras
